```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

X = np.arange(-2.0, 2.0, 0.1)
#print(type(X[0]))
X = tf.cast(X, tf.float32)

W = tf.Variable(tf.random.normal([1]), dtype = tf.float32)
b = tf.Variable(tf.random.normal([1]), dtype = tf.float32)

@tf.function
def linear(X):
  return X * W + b

@tf.function
def quadratic(X):
  return X * X + W * X + b

@tf.function
def tertiary(X):
  return X * X * X + X * X + W * X + b;

Y = linear(X)
plt.plot(X, Y, 'ro', label = '1st degree')
plt.xlim([-2.0, 2.0])
plt.ylim([0.0, 0.7])
plt.plot(X, Y)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

Y = quadratic(X)
plt.plot(X, Y, 'ro', label = '1st degree')
plt.xlim([-2.0, 2.0])
plt.ylim([0.0, 0.7])
plt.plot(X, Y)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

Y = tertiary(X)
plt.plot(X, Y, 'ro', label = '1st degree')
plt.xlim([-2.0, 2.0])
plt.ylim([0.0, 0.7])
plt.plot(X, Y)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
