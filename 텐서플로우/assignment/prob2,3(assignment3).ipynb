```python
import numpy as np
import matplotlib.pylab as plt
x = np.arange(-6, 6, 0.01)
#==========예, Sigmoid 함수 코드 =============
def sigmoid_func(x): # sigmoid 함수
  return 1 / (1 + np.exp(-x))

def arctan_func(x):
  return np.arctan(x)

def leaky_ReLU_func(x, a):
  return (x >= 0) * x + (x < 0) * a * x

def gaussian_func(x):
  return np.exp(-np.square(x))

def ELU_func(x, a):
  return (x >= 0) * x + (x < 0) * a * (np.exp(x) - 1)
    
#그래프 출력
plt.plot(x, leaky_ReLU_func(x, 0.3), linestyle='--', label="leaky_ReLU")
#========================================
plt.ylim(-5, 5)
plt.legend()
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([[1, 1], [1, 0], [0, 1], [0, 0]])
#AND
#y = np.array([[1], [0], [0], [0]])
#OR
y = np.array([[1], [1], [1], [0]])

def arctan_func(x):
  return np.arctan(x)

model = Sequential()
model.add(Dense(input_dim = 2, units = 1, activation = arctan_func))

print('-------initial weights----------')
for weight in model.weights:
  print(weight)

model.compile(optimizer = tf.optimizers.Adam(0.01),
              loss = 'categorical_crossentropy',
              metrics = ['accuracy'])
model.summary()
history = model.fit(x,y,epochs = 100, batch_size = 1)

print('----------weights----------')
for weight in model.weights:
  print(weight)

print('---------test results--------')
print(x)
print(model.predict(x))
print('\n Accuracy: %.4f' %(model.evaluate(x, y)[1]))

print('-----------------------------')
```
