```python
import tensorflow as tf

X = [[1., 2., 3.]]

W = tf.Variable(tf.random.normal([3, 1], dtype = tf.float32), name ="W")
b = tf.Variable(tf.random.normal([1,1], dtype = tf.float32), name="b")

Y = tf.matmul(X, W) + b
print("=== X ===")
print(X)
print("=== W ===")
print(W)
print("=== b ===")
print(b)
print("=== Y ===")
print(Y)
```

```python
import tensorflow as tf

X = [[1., 2., 3.],
     [4., 5., 6.]]

W = tf.Variable(tf.random.normal([3, 2], dtype = tf.float32), name ="W")
b = tf.Variable(tf.random.normal([2,1], dtype = tf.float32), name="b")

@tf.function
def forward(X) :
  return tf.matmul(X, W) + b

Y = forward(X)
print("=== X ===")
print(X)
print("=== W ===")
print(W)
print("=== b ===")
print(b)
print("=== Y ===")
print(Y)
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pylab as plt

X = [1., 2., 3.]

W = tf.Variable(tf.random.normal([1], dtype = tf.float32), name ="W")
b = tf.Variable(tf.random.normal([1], dtype = tf.float32), name="b")

@tf.function
def forward(X) :
  return X * W + b

Y = forward(X)
print(Y)

plt.plot(X, Y, 'ro')
plt.plot(X, Y)
plt.show()

plt.title("Plot")
plt.plot([1,4,9,16], 'ro')
plt.show()

plt.title("x ticks")
plt.plot([10, 20, 30,40], [1,4,9,16])
plt.show()

x = np.arange(0, 5, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

num_points = 300
vectors_set = []
for i in range(num_points):
  x = np.random.normal(0.0, 0.55)
  y = x * 0.1 + np.random.normal(0.0, 0.03) + 0.3
  vectors_set.append([x, y])

x_data = [v[0] for v in vectors_set]
y_data = [v[1] for v in vectors_set]

plt.plot(x_data, y_data, 'ro', label = 'Normal Distribution')
plt.ylim([0.1, 0.5])
plt.xlim([-2.0, 2.0])
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pylab as plt

x = tf.linspace(-3.0, 3, 1000)
y1 = tf.nn.sigmoid(x)
y2 = tf.nn.softplus(x)
y3 = tf.nn.softsign(x)
plt.plot(x, y1, "-", label = "sigmoid")
plt.plot(x, y2, "--", label = "softplus")
plt.plot(x, y3, ":", label = "softsign")
plt.xlabel("x")
plt.title("tf.nn")
plt.legend()
plt.show()
```
